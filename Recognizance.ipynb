```python
import pandas as pd  #Working with DataFrames
import numpy as np   # For Matrix Operations
import matplotlib.pyplot as plt # Visualization of the Data
import matplotlib.image as mpimg # Image formation from the pixels
import seaborn as sns # Visualization of Graphs
%matplotlib inline

np.random.seed(2)

from sklearn.model_selection import train_test_split # Spliting the data into Traing and Validation Set
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression# To check the similarity between Predictions and Answers
import itertools
```

```python
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')
```

```python
train_set.head()
```

```python
train_set = train_set.drop(['Attribute5'], axis = 1)
test_set = test_set.drop(['Attribute5'], axis=1)
```

```python
combine = [train_set,test_set]
```

```python
for x in combine:
    x['Attribute8'] = x['Attribute8'].map({'f':1,'C':2,'b':3,'a':4,'A':5 , 'B':6,'F':7,'d':8,'E':9,'D':10 ,'e':11,'c':12})
train_set.head()
```

```python
train_set.describe()
```

```python
for k in combine:
    k['Attribute1'] = k['Attribute1'].fillna(k['Attribute1'].mean())
    k['Attribute2'] = k['Attribute2'].fillna(k['Attribute2'].mean())
    k['Attribute3'] = k['Attribute3'].fillna(k['Attribute3'].mean())
    k['Attribute4'] = k['Attribute4'].fillna(k['Attribute4'].mean())
    #k['Attribute5'] = k['Attribute5'].fillna(k['Attribute5'].mean())
    k['Attribute6'] = k['Attribute6'].fillna(k['Attribute6'].mean())
    k['Attribute7'] = k['Attribute7'].fillna(k['Attribute7'].mean())
    k['Attribute8'] = k['Attribute8'].fillna(k['Attribute8'].mean())
train_set.describe()
```

```python
for dataset in combine:
    dataset['Attribute1'] = (dataset['Attribute1'] - dataset['Attribute1'].mean())/(dataset['Attribute1'].max()-dataset['Attribute1'].min())*10
    dataset['Attribute2'] = (dataset['Attribute2'] - dataset['Attribute2'].mean())/(dataset['Attribute2'].max()-dataset['Attribute2'].min())*10
    dataset['Attribute3'] = (dataset['Attribute3'] - dataset['Attribute3'].mean())/(dataset['Attribute3'].max()-dataset['Attribute3'].min())*10
    dataset['Attribute4'] = (dataset['Attribute4'] - dataset['Attribute4'].mean())/(dataset['Attribute4'].max()-dataset['Attribute4'].min())*10
    #dataset['Attribute5'] = (dataset['Attribute5'] - dataset['Attribute5'].mean())/(dataset['Attribute5'].max()-dataset['Attribute5'].min())*10
    dataset['Attribute6'] = (dataset['Attribute6'] - dataset['Attribute6'].mean())/(dataset['Attribute6'].max()-dataset['Attribute6'].min())*10
    dataset['Attribute7'] = (dataset['Attribute7'] - dataset['Attribute7'].mean())/(dataset['Attribute7'].max()-dataset['Attribute7'].min())*10
    dataset['Attribute8'] = (dataset['Attribute8'] - dataset['Attribute8'].mean())/(dataset['Attribute8'].max()-dataset['Attribute8'].min())*10
train_set.head()
```

```python
X = train_set.drop(['Pressure'],axis=1)
Y = train_set['Pressure']
print(X, X.shape , Y , Y.shape)
```
